```python
import pandas as pd
import matplotlib.pyplot as plt

# Length
def Count(x):
    count = 0
    for i in x:
        count+=1
    return count

# Sum
def Summation(x):
    total = 0
    for i in x:
        total +=i
    return total

#Mean
def Mean(x):
    return Summation(x)/Count(x)

#splits data into train and test sets
def train_test_split(x, y, split):
    size = int(Count(x)-(Count(x)*float(split)))
    x_train = x[:size] #train set
    y_train = y[:size]
    x_test = x[size:] #test set
    y_test = y[size:]
    
    return x_train, y_train, x_test, y_test

#covariance of x and y
def Covariance(x,y):
    m = Mean(x)  
    x_list = list()
    y_list = list()
    product = list()
    for i in x:
        x_list.append(i-m)
    for i in y:
        y_list.append(i-m)
    for i,j in zip(x_list, y_list):
        product.append(i*j)
    #print("Module covariance : ",Summation(product))
    return Summation(product)/(Count(x)-1)

#variance of x
def Variance(x):
    #print("Pandas Output for variance:", x.var())
    m = Mean(x)
    x_list = list()
    for i in x:
        x_list.append((i-m)**2)
    #print("Module variance : ", Summation(x_list))
    return Summation(x_list)/(Count(x)-1)

def a_value(x,y):
    return Mean(y)-(b_value(x, y)*Mean(x))

def b_value(x,y):
    return Covariance(x,y)/Variance(x)

# Y predict or Y hat
def y_Pred(x,y):
    y_hat_list = list()
    a = a_value(x,y)
    b = b_value(x,y)
    for i in x:
        y_hat_list.append((b*i)+a)
    return y_hat_list

#R-squared
def R_Squared(y,y_p):
    mn = Mean(y)
    residual = list()
    Tot_sum = list()
    for i,j in zip(y, y_p):
        residual.append((i-j)**2)
        rss=Summation(residual)
    for i in y:
        Tot_sum.append((i-mn)**2)
        tss=Summation(Tot_sum)
    return 1-(rss/tss)

#MSE(Mean Squared Error)
def MSE(y,y_p):
    error = list()
    for i,j in zip(y,y_p):
        error.append((i-j)**2)
    return (1/Count(y))*Summation(error)

#Absolute
def Absolute(series):    
    return ((series*series)**0.5);

#MAE(Mean Absolute Error)
def MAE(y,y_p):
    error = list()
    for i,j in zip(y,y_p):
        error.append(Absolute(i-j))
    return (1/Count(y))*Summation(error)


df = pd.read_csv('SimpleLinearRegression.csv')
'''df.colums = ['No', 'X1 transaction', 'X2 house age', 
             'X3 distance to the nearest MRT station',	
             'X4 number of convenience stores','X5 latitude',
             'X6 longitude','Y house price of unit area']'''


#splitting data
x = df['YearsExperience']
y = df['Salary']
print(df.head(5))
x_train, y_train, x_test, y_test = train_test_split(x, y, 1/3)


#subplot 1 for train set

plt.figure(figsize=(10,4),dpi=100)
plt.subplot(1,2,1)
ax = plt.gca()
y_pred = y_Pred(x_train, y_train)
plt.scatter(x_train, y_train, s=8)
plt.plot(x_train, y_pred, color = 'red')
plt.xlabel("YearsExperience", size=20, color='#4b0101')
plt.ylabel("Salary", size=20, color='#4b0101')
plt.title("Training Set", size=20, color='#4b0101')

#subplot 2 for test set

#plt.figure(figsize=(10,4),dpi=100)
plt.subplot(1,2,2)
ax2 = plt.gca()
y_pred = y_Pred(x_test, y_test)
plt.scatter(x_test, y_test, s=8, color='purple')
plt.plot(x_test, y_pred, color='red')
plt.xlabel("YearsExperience", size=20, color='#750851')
plt.ylabel("Salary", size=20, color='#750851')
plt.title("Test Set", size=20, color='#750851')
plt.show()

y_pred = y_Pred(x_test, y_test)
print('R-squared value: ', R_Squared(y_test, y_pred))
print('MSE value: ', MSE(y_test, y_pred))
print('MAE value: ', MAE(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#splits data into train and test sets
def train_test_split(x, y, split):
    size = int(len(x)-(len(x)*float(split)))
    x_train = x[:size] #train set
    y_train = y[:size]
    x_test = x[size:] #test set
    y_test = y[size:]
    
    return x_train, y_train, x_test, y_test

#covariance of x and y
def Covariance(x,y):
    m = np.mean(x)  
    x_list = list()
    y_list = list()
    product = list()
    for i in x:
        x_list.append(i-m)
    for i in y:
        y_list.append(i-m)
    for i,j in zip(x_list, y_list):
        product.append(i*j)
    #print("Module covariance : ",sum(product))
    return sum(product)/(len(x)-1)

#variance of x
def Variance(x):
    #print("Pandas Output for variance:", x.var())
    m = np.mean(x)
    x_list = list()
    for i in x:
        x_list.append((i-m)**2)
    #print("Module variance : ", sum(x_list))
    return sum(x_list)/(len(x)-1)

def a_value(x,y):
    return np.mean(y)-(b_value(x, y)*np.mean(x))

def b_value(x,y):
    return Covariance(x,y)/Variance(x)

# Y predict or Y hat
def y_Pred(x,y):
    y_hat_list = list()
    a = a_value(x,y)
    b = b_value(x,y)
    for i in x:
        y_hat_list.append((b*i)+a)
    return y_hat_list

#R-squared
def R_Squared(y,y_p):
    mn = np.mean(y)
    residual = list()
    Tot_sum = list()
    for i,j in zip(y, y_p):
        residual.append((i-j)**2)
        rss=sum(residual)
    for i in y:
        Tot_sum.append((i-mn)**2)
        tss=sum(Tot_sum)
    return 1-(rss/tss)

#MSE(np.mean Squared Error)
def MSE(y,y_p):
    error = list()
    for i,j in zip(y,y_p):
        error.append((i-j)**2)
    return (1/len(y))*sum(error)

#Absolute
def Absolute(series):    
    return ((series*series)**0.5);

#MAE(np.mean Absolute Error)
def MAE(y,y_p):
    error = list()
    for i,j in zip(y,y_p):
        error.append(Absolute(i-j))
    return (1/len(y))*sum(error)


df = pd.read_csv('SimpleLinearRegression.csv')
'''df.colums = ['No', 'X1 transaction', 'X2 house age', 
             'X3 distance to the nearest MRT station',	
             'X4 number of convenience stores','X5 latitude',
             'X6 longitude','Y house price of unit area']'''


#splitting data
x = df['YearsExperience']
y = df['Salary']
print(df.head(5))
x_train, y_train, x_test, y_test = train_test_split(x, y, 1/3)


#subplot 1 for train set

plt.figure(figsize=(10,4),dpi=100)
plt.subplot(1,2,1)
ax = plt.gca()
y_pred = y_Pred(x_train, y_train)
plt.scatter(x_train, y_train, s=8)
plt.plot(x_train, y_pred, color = 'red')
plt.xlabel("YearsExperience", size=20, color='#4b0101')
plt.ylabel("Salary", size=20, color='#4b0101')
plt.title("Training Set", size=20, color='#4b0101')

#subplot 2 for test set

#plt.figure(figsize=(10,4),dpi=100)
plt.subplot(1,2,2)
ax2 = plt.gca()
y_pred = y_Pred(x_test, y_test)
plt.scatter(x_test, y_test, s=8, color='purple')
plt.plot(x_test, y_pred, color='red')
plt.xlabel("YearsExperience", size=20, color='#750851')
plt.ylabel("Salary", size=20, color='#750851')
plt.title("Test Set", size=20, color='#750851')
plt.show()

y_pred = y_Pred(x_test, y_test)
print('R-squared value: ', R_Squared(y_test, y_pred))
print('MSE value: ', MSE(y_test, y_pred))
print('MAE value: ', MAE(y_test, y_pred))
```
