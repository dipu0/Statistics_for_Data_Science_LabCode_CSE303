```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator

# Length
def Count(x):
    count = 0
    for i in x:
        count+=1
    return count

# Sum
def Summation(x):
    total = 0
    for i in x:
        total +=i
    return total

#Mean
def Mean(x):
    return Summation(x)/Count(x)

#R-squared
def R_Squared(y,y_p):
    mn = Mean(y)
    residual = list()
    Tot_sum = list()
    for i,j in zip(y, y_p):
        residual.append((i-j)**2)
        rss=Summation(residual)
    for i in y:
        Tot_sum.append((i-mn)**2)
        tss=Summation(Tot_sum)
    return 1-(rss/tss)

#MSE(Mean Squared Error)
def MSE(y,y_p):
    error = list()
    for i,j in zip(y,y_p):
        error.append((i-j)**2)
    return (1/Count(y))*Summation(error)

#Absolute
def Absolute(series):    
    return ((series*series)**0.5);

#MAE(Mean Absolute Error)
def MAE(y,y_p):
    error = list()
    for i,j in zip(y,y_p):
        error.append(Absolute(i-j))
    return (1/Count(y))*Summation(error)

def Dot(v1,v2):
    result=[[Summation(a*b for a,b in zip(X_row,Y_col)) for Y_col in zip(*v2)] 
            for X_row in v1]
    return result

def Transpose(matrix):
    rows = Count(matrix)
    columns = Count(matrix[0])
    trans_mat=[]
    for j in range(columns):
        row = []
        for i in range(rows):
           row.append(matrix[i][j])
        trans_mat.append(row)
    return trans_mat

def Weights(x,y):  
    tp=Transpose(x)
    dot1=Dot(tp,x)
    length= Count(dot1)
    print("Length of matrix is: ", length)
    #det = Determinant(dot1,length)
    #det = np.linalg.det(x)
    inverse= np.linalg.inv(dot1)
    dot2= Dot(tp,y)
    weight= Dot(inverse,dot2)
    print("Weight :",weight)
    return weight



df = pd.read_csv('RealEstate.csv')
'''df.colums = ['No', 'X1 transaction', 'X2 house age', 
             'X3 distance to the nearest MRT station',	
             'X4 number of convenience stores','X5 latitude',
             'X6 longitude','Y house price of unit area']'''


df['bias']=1
print(df.head(5))

x = df[['bias', 'X2 house age', 'X6 longitude']]
y = df[['Y house price of unit area']]


x = np.array(x)
y = np.array(y)


w=Weights(x,y)
y_pred =w[0]+w[1]*df['X2 house age']+w[2]*df["X6 longitude"]


fig = plt.figure()
fig,ax=plt.subplots(subplot_kw={"projection": "3d"})
ax = plt.gca()
ax.set_facecolor('#EAE0DE')
surf= ax.scatter(x[:,0],x[:,1],y,cmap=cm.coolwarm,linewidth=0, color= 'red',antialiased=False)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter('{x:.02f}')
plt.title("Multiple Linear Regression")
plt.show()

print('R-squared value: ', R_Squared(y, y_pred))
print('MSE value: ', MSE(y, y_pred))
print('MAE value: ', MAE(y, y_pred))
```
